```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('heart failure.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.rename({'DEATH_EVENT':'death'},axis=1)
df
```

```python
x = df.drop('death',axis=1)
y = df['death']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain, xtest, ytrain, ytest = tts(x,y,train_size=.70,random_state=2021)
```

```python
ytest
```

```python
ytest.shape
```

# Apply Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(xtrain,ytrain)
```

```python
pred = dt.predict(xtest)
```

```python
ytest
```

```python
pred 
```

```python
dt.score(xtest,ytest)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier( random_state=2)
```

```python
rfc.fit(xtrain, ytrain)
```

```python
rfc.score(xtest, ytest)
```

# Grid Search CV

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#grids
ids = { 
    'n_estimators': np.random.randint(250,500,20),
    'max_features': ['sqrt', 'log2'],
    'max_depth' : np.random.randint(25,50,20),
    'criterion' :['gini', 'entropy']
} 
```

```python
gsc = GridSearchCV(estimator=rfc, param_grid = ids)
```

```python
gsc.fit(xtrain,ytrain)
```

```python
gsc.score(xtest, ytest)
```

```python
gsc.cv_results_
```

```python
gsc_cv_result = pd.DataFrame(gsc.cv_results_)
gsc_cv_result
```

```python
gsc.best_score_
```

```python
gsc.best_params_
```

```python
gsc.best_estimator_
```

```python
rfc2 = gsc.best_estimator_
rfc2
```

```python
rfc2.fit(xtrain, ytrain)
```

```python
rfc2.score(xtest,ytest)
```
